```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

iris_new = iris.copy()

iris_new['species'] = le.fit_transform(iris_new['species'])
iris_new['species'].unique()
```

```python
iris_new.columns
```

```python
x_data = range(0, iris.shape[0])
columns = list(iris_new.columns)

fig, ax = plt.subplots()

#PLOT EACH COLUMN

for i in columns:
    ax.plot(x_data, iris_new[i], label= i)

#let's make the y axis have no ticks
ax.yaxis.set_ticklabels([])
#let's give y axis a label instead
ax.set_ylabel('Iris columns')

ax.set_title('Iris Dataset')
ax.legend()
```

```python
x_data = range(0, iris.shape[0])
columns = list(iris_new.columns)

fig, ax = plt.subplots(figsize= (20, 5))

#PLOT EACH COLUMN

for i in columns:
    ax.plot(x_data, iris_new[i], label= i)

#let's make the y axis have no ticks
ax.yaxis.set_ticklabels([])
#let's give y axis a label instead
ax.set_ylabel('Iris columns')

ax.set_title('Iris Dataset')
ax.legend()
```
