For Wine dataset

1. Do a scattered plot with marplotlib, pick any column for x and y
2. Do a histogram of all columns using pandas 
3. Show a horizontal bar chart of any column using seaborn 
4. Show the line chart of all columns using marplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
wine = pd.read_csv('wine.csv')
iris = pd.read_csv('iris.csv')

```

```python
wine.head()
```

```python
# 1.

plt.scatter(wine['residual sugar'], wine['alcohol'], label='sugar against alcohol')
plt.title('wine')
plt.xlabel('sugar')
plt.ylabel('alcohol')
plt.legend()
```

```python
# 1
fig, ax = plt.subplots()

ax.scatter(wine['residual sugar'], wine['alcohol'], label='sugar against alcohol')
ax.set_title('wine')
ax.set_xlabel('sugar')
ax.set_ylabel('alcohol')
#plt.plot(wine)
plt.legend()
```

```python
# 2.

wine.hist(figsize=(10, 12))
```

```python
# 3

sns.countplot(y ='quality', data= wine)
```

```python
# 4

x_data = range(0, wine.shape[0])
columns = list(wine.columns)

fig, ax = plt.subplots(figsize=(10, 15))

#PLOT EACH COLUMN

for i in columns:
    ax.plot(x_data, wine[i], label= i)

#let's make the y axis have no ticks
ax.yaxis.set_ticklabels([])
#let's give y axis a label instead
ax.set_ylabel('Wine columns')

ax.set_title('Wine Dataset')
ax.legend()
```

```python
col = ['chlorides', 'total sulfur dioxide', \
       'free sulfur dioxide', 'citric acid']
wine_new = wine.drop(col, axis=1)
```

```python
# 4

x_data = range(0, wine_new.shape[0])
columns = list(wine_new.columns)

fig, ax = plt.subplots(figsize=(20, 10))

#PLOT EACH COLUMN

for i in columns:
    ax.plot(x_data, np.log(wine_new[i]), label= i)

#let's make the y axis have no ticks
ax.yaxis.set_ticklabels([])
#let's give y axis a label instead
ax.set_ylabel('Wine columns')

ax.set_title('Wine Dataset')
ax.legend()
```

```python
# 4

x_data = range(0, wine_new.shape[0])
columns = list(wine_new.columns)

#PLOT EACH COLUMN

for i in columns:
    plt.plot(x_data, np.log(wine_new[i]), label= i)
    
#let's give y axis a label instead
plt.ylabel('Wine columns')

plt.title('Wine Dataset')
plt.legend()
```

For iris dataset

1. Show a scatter plot of sepal length and sepal width, where the species of each row is represented with a colour
rows with setosa should have red, 
rows with versicolor should have green,
rows with virginica should have blue

```python
iris.head()
```

```python
#with matplotlib.pyplot

colors = {'setosa':'r', 'versicolor':'g', 'virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(iris.shape[0]):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],\
               color=colors[iris['species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

```
